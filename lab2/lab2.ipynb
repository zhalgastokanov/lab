## 1.Исправить замечания преподавателя по ЛР №1.

## 3.Закодировать категориальные признаки (не менее 2 методов сравнения).

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

cat_features = ['Gender', 'Occupation', 'Device_Type']

df_ordinal = df.copy()
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
df_ordinal[cat_features] = ordinal_encoder.fit_transform(df_ordinal[cat_features])

df_onehot = df.copy()
onehot_encoder = OneHotEncoder(sparse_output=False, drop='first')  # <-- изменено
encoded = pd.DataFrame(onehot_encoder.fit_transform(df_onehot[cat_features]),
                       columns=onehot_encoder.get_feature_names_out(cat_features),
                       index=df_onehot.index)
df_onehot = pd.concat([df_onehot.drop(cat_features, axis=1), encoded], axis=1)

print("Пример Ordinal Encoding:")
display(df_ordinal.head())

print("Пример One-Hot Encoding:")
display(df_onehot.head())
```

## 4.Выполнить масштабирование числовых признаков.

```python
from sklearn.preprocessing import StandardScaler

num_features = ['Age', 'Daily_Phone_Hours', 'Social_Media_Hours', 'Work_Productivity_Score',
                'Sleep_Hours', 'Stress_Level', 'App_Usage_Count', 'Caffeine_Intake_Cups', 
                'Weekend_Screen_Time_Hours']

def scale_features(df, columns):
    scaler = StandardScaler()
    df[columns] = scaler.fit_transform(df[columns])
    return df

df_ordinal_scaled = scale_features(df_ordinal.copy(), num_features)
df_onehot_scaled = scale_features(df_onehot.copy(), num_features)

print("Ordinal Encoding + масштабирование:")
display(df_ordinal_scaled.head())

print("One-Hot Encoding + масштабирование:")
display(df_onehot_scaled.head())
```

## 5.Создать дополнительные признаки (минимум 2 новых).

```python
for df_scaled in [df_ordinal_scaled, df_onehot_scaled]:
    df_scaled['Total_Screen_Time'] = df_scaled['Daily_Phone_Hours'] + df_scaled['Weekend_Screen_Time_Hours']
    df_scaled['Productivity_per_Sleep'] = df_scaled['Work_Productivity_Score'] / df_scaled['Sleep_Hours']

print("Ordinal Encoding + новые признаки:")
display(df_ordinal_scaled.head())

print("One-Hot Encoding + новые признаки:")
display(df_onehot_scaled.head())
```

## 6.Удалить нерелевантные признаки (обосновать).
**В датасете есть признак:**

- User_ID — уникальный идентификатор пользователя.

Обоснование:

- Он не несет полезной информации для обучения модели, так как каждое значение уникально;

- Включение такого признака может только ухудшить обучение, так как модель будет пытаться подгонять данные под конкретные ID (overfitting);

- Все остальные признаки несут количественную или категориальную информацию, полезную для анализа и предсказания.

```python
irrelevant_features = ['User_ID']

df_ordinal_final = df_ordinal_scaled.drop(irrelevant_features, axis=1)
df_onehot_final = df_onehot_scaled.drop(irrelevant_features, axis=1)

print("Ordinal Encoding — финальный датасет:")
display(df_ordinal_final.head())

print("One-Hot Encoding — финальный датасет:")
display(df_onehot_final.head())
```

## 7.Подготовить финальный датасет для модели.


```python
df_ordinal_final.to_csv("dataset_final_ordinal.csv", index=False)
df_onehot_final.to_csv("dataset_final_onehot.csv", index=False)

print("Финальные датасеты сохранены:")
print("- Ordinal Encoding: dataset_final_ordinal.csv")
print("- One-Hot Encoding: dataset_final_onehot.csv")
```
